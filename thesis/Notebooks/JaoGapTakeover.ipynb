```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from scipy.interpolate import interp1d
plt.rcParams["text.usetex"]  = False
```

```python
arxiv = pd.read_csv("get_monthly_submissions.csv")
arxiv['month'] = pd.to_datetime(arxiv['month'])
```

```python
subs = arxiv.groupby((arxiv['month'].dt.year//10)*10)['submissions'].sum()
```

```python
jCpapers = [0, 0, 2, 11]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
X = subs.index.values
Y = [x/y for x, y in zip(jCpapers, subs.values)]
ax.plot(X, Y , 'ko--')
ax.set_ylabel("Fraction of Papers on the ArXiv", fontsize=20)
ax.set_xlabel("Year [yr]", fontsize=25)
fig.savefig("Figures/CurrentPapers.pdf")
```
